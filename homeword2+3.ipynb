```python
#bài8
import numpy as np

# Nhập hệ số
a, b, c, d, m, n = map(float, input("Nhập a, b, c, d, m, n: ").split())

# Tạo ma trận
A = np.array([[a, b],
              [c, d]])
B = np.array([m, n])

try:
    x, y = np.linalg.solve(A, B)
    print(f"Hệ có nghiệm duy nhất: x = {x}, y = {y}")
except np.linalg.LinAlgError:
    print("Hệ vô nghiệm hoặc vô số nghiệm")
```
