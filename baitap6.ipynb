```python
#9
# Nhập dict dạng: a:1 b:2
def read_dict():
    d = {}
    for pair in input().split():
        k, v = pair.split(':')
        d[k] = int(v)
    return d

d1 = read_dict()
d2 = read_dict()

result = {}

for k in d1:
    result[k] = result.get(k, 0) + d1[k]

for k in d2:
    result[k] = result.get(k, 0) + d2[k]

for k in sorted(result):
    print(k, result[k])
```

```python
#10
def w6a10(lst, k):
    # Sử dụng set để lưu các cặp số nhằm loại bỏ trùng lặp
    pairs = set()
    n = len(lst)
    
    # Duyệt qua từng phần tử và các phần tử đứng sau nó
    for i in range(n):
        for j in range(i + 1, n):
            if lst[i] + lst[j] == k:
                pairs.add((lst[i], lst[j]))
    
    # Chuyển về list và sắp xếp theo phần tử đầu tiên của tuple
    result = sorted(list(pairs), key=lambda x: x[0])
    print(result)

```

```python
#11

def w6a11(input_tuple):
    # Dùng list comprehension (hoặc generator expression) để lọc
    even_nums = tuple(x for x in input_tuple if x % 2 == 0)
    odd_nums = tuple(x for x in input_tuple if x % 2 != 0)
    
    print(even_nums)
    print(odd_nums)

```

```python
#12
def w6a12(lst):
    if not lst: return None
    
    # Đếm tần suất xuất hiện
    counts = {}
    for num in lst:
        counts[num] = counts.get(num, 0) + 1
    
    # Tìm tần suất lớn nhất
    max_freq = max(counts.values())
    
    # Lọc ra các số có tần suất bằng max_freq
    candidates = [num for num, freq in counts.items() if freq == max_freq]
    
    # In ra số nhỏ nhất trong các ứng viên
    print(min(candidates))

```

```python
#14
def w6a14(list1, list2):
    # Chuyển đổi list thành set để loại bỏ trùng lặp và dễ tìm giao điểm
    set1 = set(list1)
    set2 = set(list2)
    
    # Tìm giao của 2 set (các phần tử xuất hiện ở cả 2 nơi)
    common_elements = set1 & set2  # Hoặc dùng set1.intersection(set2)
    
    # Chuyển ngược lại thành list và sắp xếp (tùy chọn, để kết quả đẹp hơn)
    result = sorted(list(common_elements))
    
    print(result)
```
