```python
#W4A6
def la_nguyen_to(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

while True:
    try:
        a, b = map(int, input("Nhập hai số a b (cách nhau bởi dấu cách, hoặc Ctrl+C để dừng): ").split())
        if a <= 0 or b <= 0 or a > b:
            print("Giá trị không hợp lệ, vui lòng nhập lại!")
            continue

        tong = 0
        for i in range(a, b + 1):
            if la_nguyen_to(i):
                tong += i

        print("Tổng các số nguyên tố trong đoạn [{}, {}] là: {}".format(a, b, tong))
    except:
        break

```

```python
#W4A7
n = int(input("Nhập số nguyên dương n (n ≥ 2): "))

if n >= 2:
    uoc_lon_nhat = 1
    i = 2
    while i <= n:
        if n % i == 0:
            uoc_lon_nhat = i
            while n % i == 0:
                n //= i
        i += 1
    print("Ước số nguyên tố lớn nhất là:", uoc_lon_nhat)
else:
    print("Giá trị n không hợp lệ!")
```

```python
#W4A8
n = int(input("Nhập số nguyên dương n (n ≥ 2): "))

if n >= 2:
    i = 2
    while i * i <= n:
        if n % i == 0:
            n //= i
        else:
            i += 1
    print("Ước số nguyên tố lớn nhất là:", n)
else:
    print("Giá trị n không hợp lệ!")
```

```python
#W4A9
n = int(input("Nhập số nguyên dương n: "))

def khac_nhau(x):
    digits = []
    while x > 0:
        d = x % 10
        if d in digits:
            return False
        digits.append(d)
        x //= 10
    return True

ds = []

for i in range(1, int(n**0.5) + 1):
    so_cp = i * i
    if khac_nhau(so_cp):
        ds.append(so_cp)

if ds:
    print(*ds)
else:
    print("no number")
```

```python
#W4A10
n = int(input("Nhập số nguyên dương n: "))

def do_dai_collatz(x):
    dem = 1
    while x != 1:
        if x % 2 == 0:
            x //= 2
        else:
            x = 3 * x + 1
        dem += 1
    return dem

max_len = 0
so_x = 1

for i in range(1, n + 1):
    L = do_dai_collatz(i)
    if L > max_len or (L == max_len and i < so_x):
        max_len = L
        so_x = i

print(so_x, max_len)
```

```python
#W4A15
# Nhập tổng số con và tổng số chân
tong_con = int(input("Nhập tổng số con: "))
tong_chan = int(input("Nhập tổng số chân: "))

tim_thay = False

for ga in range(tong_con + 1):
    cho = tong_con - ga
    if 2 * ga + 4 * cho == tong_chan:
        print("Gà:", ga, ", Chó:", cho)
        tim_thay = True
        break

if not tim_thay:
    print("invalid")
```
